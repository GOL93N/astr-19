```python
'''Create a Jupyter notebook, |import matplotlib and numpy.| Write cells that create an array x ranging from [0,2pi] 
in 100 steps and that defines a function that returns a*exp(-b*x).  Choose values for the variables a and b. In a 
new cell use the function to set y=a*exp(-b*x), and then plot x vs. y. Label your axes and save the figure as a PDF.'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)

a = 3
b = 7

def exponential(x, a, b):
    return a*np.exp(-b*x)
```

```python
y = exponential(x, a, b)
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.savefig("CJ2_P1.png")
```
