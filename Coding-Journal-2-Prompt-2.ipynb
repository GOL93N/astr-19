'''Using np.random.normal(), add scatter to your dataset. Do not give a constant sigma to every datapoint.
Instead, make it so that the uncertainty in y decreases linearly as x increases, going from 2 to 0.2. Add 
this scatter to your y variable and plot it with plt.errorbar(). Make sure to use a different color and fmt 
than the default (look in the documentation for how you can change this). Label your axes and save the figure 
as a PDF.'''

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 100)

a = 3
b = 7

def exponential(x, a, b):
    return a*np.exp(-b*x)

y = exponential(x, a, b)
```

```python
ammount= np.linspace(2, 0.2, 100)
y_scatter = y + np.random.normal(0, ammount)
```

```python
plt.errorbar(x, y_scatter, yerr=ammount, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.savefig("Journal2Prompt2.pdf")
```
